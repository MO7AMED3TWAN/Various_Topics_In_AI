```python
#### IMPORTING LIBRARIES ####
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#### READ DATA ####
path = r"test1.csv "
data = pd.read_csv(path)
data.isnull().sum() ## Get Nulls in My Data
```

# DEFINE DATA

```python
#### define features and target and Theta ####

X=data.iloc[:,:-1]
y=data.iloc[:,-1]
    
# data.columns=['','','','','',...]#تعريف اسماء الاعمدة 

#### SPLITTING DATA ####
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

# SHOWING DATA


```python
print(f"That's Information About My Data \n{data.info()}\n\n")

print(f"That's My First 5 Rows IN My Data \n{data.head()}\n\n")

print(f"That's The Describtion Of My Data \n{data.describe()}\n\n")

print(f"SCATTER OF MY DATA \n{plt.scatter(data.X,data.Y ,c='r' )}")
```

# DRAWING BEST FIT LINE 


```python
sns.regplot(x='X',y='Y',data=data,marker='+')
```

# CREATE A MODEL OF PREDECTION


```python
### MAKING MODEL ###
L_G_MODEL=LinearRegression()

### TRAINING MODEL ###
L_G_MODEL.fit(X_train,y_train)

### PREDECT BY MODEL ###
Predect_Values=L_G_MODEL.predict(X_test)
```

## DRAW SCATTER OF RESULTS

```python
#Draw Scatter
fig, shape = plt.subplots(figsize=(5,5))
shape.scatter(X_train,y_train,c='b'        , label='Original Data')
shape.scatter(X_test,Predect_Values,c='r', label='Prediction')
shape.legend(loc=0)
#describ the shape
shape.set_xlabel('X')
shape.set_ylabel('Y & Predect')
shape.set_title('MY ORIGINAL & PREDECT DATA')
plt.show()
```

## The Coefficients

```python
print("Coefficients: \n", L_G_MODEL.coef_)
```

## SCORE OF ERROR

```python
#Import Libraries
from sklearn.metrics import mean_squared_error 
#----------------------------------------------------
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, Predect_Values) # it can be raw_values
print('Mean Squared Error Value is : ',round(MSEValue,4))
```

```python
#Import Libraries
from sklearn.metrics import r2_score
#----------------------------------------------------
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test,Predect_Values))
```
